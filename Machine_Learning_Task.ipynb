# MNIST Digits Summation

image.png

**iconicchain** has certain services that utilize Computer Vision models. Therefore this task is created to assess the knowledge of data augmentation, data processing, modeling, and evaluation of machine learning algorithms related to the Computer Vision tasks. 

The goal of this task is to create a neural network architecture that **receives as input two handwritten digit images** and **provides their sum as the output**. The summation logic should be learned by the model itself, instead of predicting the digits from images and then summing them together explicitly. 

image.png

# Task Description:
The task is composed of the following steps:<br>
**- Download the MNIST dataset**<br>
**- Prepare and preprocess the data**<br>
  - Use some visualization to understand the input examples and their labels.

**- Increase the size of the dataset with data augmentation techniques**<br>
   - We might not know the source application that produces this handwritten images in prior. How to transform the dataset to make the model invariant to rotations/translations/scaling to some extent? <br>
📝 The inputs will always be binary images with nothing else (except noise) in the frame.<br>

**- Train the model for calculating the sum of 2 digits**<br>
  - Use the appropriate metrics to monitor training performance
  
**- Evaluate the model's performance on the test dataset**<br>
  - Generally, the MNIST classification models can achieve quite high accuracy. Can you write a conclusion comparing your accuracy with those models and why is your model performing similar or not similar by looking at the predictions?
  - How well does the model perform for your own inputs?

Decisions made during each of the above steps needs to be documented and elaborated. The machine learning framework that is preferable for training the model is PyTorch. Good luck!
